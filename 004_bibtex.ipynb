```python
from os.path import join
import pandas as pd
import numpy as np
import requests
import json
```

```python
df = pd.read_csv("dr_algorithms_with_ids.csv", index_col=0)
```

```python
acronyms = df['Acronym'].tolist()
```

```python
ss_ids = df['ss_id'].tolist()
```

```python
r = requests.post(
    'https://api.semanticscholar.org/graph/v1/paper/batch',
    params={'fields': 'authors,year,title,citationStyles'},
    json={"ids": ss_ids}
)
```

```python
bibtex = [f"% {a}\n" + v["citationStyles"]["bibtex"] for a, v in zip(acronyms, r.json())]
```

```python
print("\n".join(bibtex))
```
